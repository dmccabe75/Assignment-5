```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "Assignment-5/city_data.csv"
ride_data_to_load = "Assignment-5/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv('city_data.csv')
ride_data = pd.read_csv('ride_data.csv')

# Combine the data into a single dataset
combined_data_df = pd.merge(city_data, ride_data, on="city")
combined_data_df.head()
# Display the data table for preview
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
#citytypes_df = combined_data_df.groupby(["type", "city"])
urban_df = combined_data_df.set_index("type")
urban_df = urban_df.loc["Urban"]
urban_df = urban_df.groupby("city")
urban_avg_fare = urban_df["fare"].mean()
urban_total_rides = urban_df["ride_id"].count()
urban_driver_count = city_data.set_index("type").loc["Urban"]
urban_driver_count = urban_driver_count["driver_count"]
urban_driver_count

# Build the scatter plots for each city types
urban_x = urban_total_rides
urban_y = urban_avg_fare

area = (urban_driver_count)**1.75  # 0 to 15 point radii

plt.scatter(urban_x, urban_y, c='gold', edgecolors="black", marker="o", s=(area), alpha=.75)

urban_df = combined_data_df.set_index("type")
suburban_df = urban_df.loc["Suburban"]
suburban_df = suburban_df.groupby("city")
suburban_avg_fare = suburban_df["fare"].mean()
suburban_total_rides = suburban_df["ride_id"].count()
suburban_driver_count = city_data.set_index("type").loc["Suburban"]
suburban_driver_count = suburban_driver_count["driver_count"]
suburban_driver_count

# Build the scatter plots for each city types
suburban_x = suburban_total_rides
suburban_y = suburban_avg_fare

area = (suburban_driver_count)**1.75  # 0 to 15 point radii

plt.scatter(suburban_x, suburban_y, c='skyblue',edgecolors="black", marker="o", s=(area), alpha=.75)

urban_df = combined_data_df.set_index("type")
rural_df = urban_df.loc["Rural"]
rural_df = rural_df.groupby("city")
rural_avg_fare = rural_df["fare"].mean()
rural_total_rides = rural_df["ride_id"].count()
rural_driver_count = city_data.set_index("type").loc["Rural"]
rural_driver_count = rural_driver_count["driver_count"]
rural_driver_count

# Build the scatter plots for each city types
rural_x = rural_total_rides
rural_y = rural_avg_fare

area = (rural_driver_count)**1.75  # 0 to 15 point radii

plt.scatter(rural_x, rural_y, c='coral', edgecolors="black", marker="o", s=(area), alpha=.75)


# # # Incorporate the other graph properties
plt.grid(b=None, which='major', axis='both')
plt.title("Pyber Ride Sharing Data (2016)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.legend(loc="best", labels= ["Urban", "Suburban", "Rural"], title = "City Type", markerscale=.5)
plt.text(60, 40, "Note: Circle size correlates with driver count per city")

plt.tight_layout()
plt.show()

plt.savefig("../Assignment-5/PyberScatter.png")

```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
fare_df = combined_data_df.groupby(["type"])
fare_breakdown = fare_df["fare"].sum()
fare_breakdown
# Build Pie Chart
labels = ("Rural", "Suburban", "Urban")
colors = ("gold", "skyblue", "coral")
explode = (0,0,.3)
plt.title("% of Total Fares by City Type")
plt.pie(fare_breakdown, colors=colors, explode=explode, labels=labels,
        autopct="%1.1f%%", radius=5, pctdistance = .65, shadow=True, startangle=150)
plt.figsize=(5, 3)
plt.axis("equal")
plt.show()

plt.savefig("../Assignment-5/FaresPie.png")
# Save Figure
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
fare_df = combined_data_df.groupby(["type"])
ride_breakdown = fare_df["ride_id"].count()
ride_breakdown
# Build Pie Chart
labels = ("Rural", "Suburban", "Urban")
colors = ("gold", "skyblue", "coral")
explode = (0,0,0.4)
plt.title("% of Total Rides by City Type")
plt.pie(ride_breakdown, colors=colors, explode=explode, labels=labels,
        autopct="%1.1f%%", shadow=True, radius=5, pctdistance = .65, startangle=140)
plt.figsize=(10, 3)
plt.axis("equal")
plt.show()

plt.savefig("../Assignment-5/RidesPie.png")

# Build Pie Chart

# Save Figure
```
